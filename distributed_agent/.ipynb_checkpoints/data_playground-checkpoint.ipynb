```python
def transform_time(x):
    time_str = str(x)
    if len(time_str)<6:
        time_str = "0{0}".format(time_str)
    splitat = 2
    h, m, s = time_str[:splitat], time_str[splitat:splitat+2], time_str[splitat+2:]
    time = "{0}:{1}:{2}".format(h, m, s)
    time_str = datetime.strptime(time, '%H:%M:%S')
    return time_str
```

```python
def split_date(x):
    date_str = str(x)
    splitat = 4
    splitto = 6
    Y, M, D = date_str[:splitat], date_str[splitat:splitto], date_str[splitto:]
    date = "{0}/{1}/{2}".format(Y, M, D)
    date_str = datetime.strptime(date, "%Y/%m/%d")
    return date_str
```

```python
def make_dictionary(pos_df_selected, df_selected):
    from datetime import datetime, timedelta
    transaction_dict = {}
    for tran_index, transaction in pos_df_selected.iterrows() :
        #print (transaction.hour) # using pandas series to get the specific columns
        #print(tran_index)
        #print(type(transaction.time))
        for row_index, row in df_selected.iterrows() :

            new_time_entry = row.ENTRY_TIME + timedelta(hours=0,minutes=5, seconds=0.5)
            new_time_exit = row.EXIT_TIME - timedelta(hours=0,minutes=5, seconds=0.5)

            if new_time_entry <= transaction.datetime <= new_time_exit :
                #print(tran_index, row_index)
                if ('%s'%tran_index not in transaction_dict) and (row_index not in transaction_dict.values() ) :
                    transaction_dict["%s"%tran_index] = [row_index]
                elif ('%s'%tran_index in transaction_dict) and (row_index not in transaction_dict.values() ) :
                    transaction_dict["%s"%tran_index].append(row_index)
                else :    
                    continue
            else :
                continue
    return transaction_dict
```

```python
def find_cars(dictionary, df_selected):
    unique_trans_list = []
    for i_list in dictionary.values():
        #print(i_list)
        for car_tran in i_list :
            if car_tran not in unique_trans_list :
                unique_trans_list.append(car_tran)
            else :
                continue
    car_list = df_selected[df_selected.index.isin(unique_trans_list)]['VEHICLE_ID'].unique()
    return car_list
```
